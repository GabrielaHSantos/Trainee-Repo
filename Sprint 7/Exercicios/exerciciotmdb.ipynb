### Definindo a chave do tmdb e a URL da API para obter a lista de filmes mais bem avaliados

```python
import requests

api_key = "0b42a987ae8b0734209304713d27f937"
url = f"https://api.themoviedb.org/3/movie/top_rated?api_key={api_key}&language=pt-BR"

```

## Fazendo a requisição e extraindo os dados dos filmes

```python
response = requests.get(url)
data = response.json()

filmes = []
for movie in data['results']:
    filmes.append({
        'Título': movie['title'],
        'Data de lançamento': movie['release_date'],
        'Visão geral': movie['overview'],
        'Votos': movie['vote_count'],
        'Média de votos': movie['vote_average']
    })
```

## Criando o DataFrame

```python
df = pd.DataFrame(filmes)
display(df)
```
